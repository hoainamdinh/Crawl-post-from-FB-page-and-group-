INITIALIZE LIST TO STORE DATA

```python
id_list = []
contributor_list = []
html = []
text_data_list = []
images_link = []
```

GET POST / CAPTION / IMAGE

```python
## find post
posts = driver.find_elements(By.CSS_SELECTOR, "div.x1yztbdb.x1n2onr6.xh8yej3.x1ja2u2z")
posts = posts[1:300]
for post in posts:
    driver.execute_script("arguments[0].scrollIntoView(true);", post)
    post_html = post.get_attribute("outerHTML")
    soup = BeautifulSoup(post_html, 'html.parser')
## get post id
    id_div = soup.find("div", class_="x6s0dn4 x3nfvp2")
    post_id_tag = soup.find_all('a', href=True)
    href = post_id_tag[-1]['href']
    match = re.search(r'/posts/([^/?]+)', href)
    if match:
        post_id = match.group(1)
        print(post_id)
    
## get contributor name
    contributor = soup.find("span", class_="html-span xdj266r x11i5rnm xat24cr x1mh8g0r xexx8yu x4uap5 x18d9i69 xkhd6sd x1hl2dhg x16tdsg8 x1vvkbs").text
## get text
    text_wrapper = soup.find_all("div", class_="html-div xdj266r x11i5rnm xat24cr x1mh8g0r xexx8yu x4uap5 x18d9i69 xkhd6sd")[4]
    text_line = text_wrapper.find_all('div', attrs={'dir': 'auto', 'style': 'text-align: start;'})
    text_data = ""
    for line in text_line:
        text = line.get_text(strip=True)
        if text:  # Chỉ thêm vào danh sách nếu có văn bản
            text_data += text + "\n" 
## get images
    all_img = soup.find_all("div", class_="html-div xdj266r x11i5rnm xat24cr x1mh8g0r xexx8yu x4uap5 x18d9i69 xkhd6sd x1n2onr6")[1]
    images = [img['src'] for img in all_img.find_all('img', src=True) if img['src'].startswith('https://scontent')]


    # Add to list
    id_list.append(post_id)
    contributor_list.append(contributor)
    html.append(post_html)
    text_data_list.append(text_data)
    images_link.append(images)

    # Break
    if len(id_list) >= 300:
        print("Đã lấy xong")
        break
    else:
        print("Đã lấy được data của post thứ", len(id_list))

```

```python
driver.quit()
```

```python
df = pd.DataFrame({
    'ID': id_list,
    'Contributor': contributor,
    'HTML': html,
    'Text': text_data,
    'Image_Link': images_link
})
```

DOWNLOAD IMAGES

```python
image_folder = 'Flex tới hơi thở cuối cùng'
os.mkdir(image_folder)
i = 0
# Iterate over the DataFrame rows
for index, row in df.iterrows():
    post_id = row['ID']
    image_links = row['Image_Link']  # This is a list of image URLs
    i += 1
    if image_links:  # Check if the list is not empty
        for j, img_url in enumerate(image_links):
            img_data = requests.get(img_url).content
            img_filename = os.path.join(image_folder, f'post_{i}_img_{j+1}.jpg')
            with open(img_filename, 'wb') as img_file:
                img_file.write(img_data)
            print(f"Đã tải ảnh {j+1} post {i}")
```
