IMPORT

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import requests
import os
import numpy as np
import pandas as pd 
import re
```

OPEN DRIVER

```python
driver = webdriver.Chrome()
page_url = "https://www.facebook.com/memetuoiturungpacbo"
driver.get(page_url)
time.sleep(3)
```

INITIALIZE LIST TO STORE DATA

```python
# initialize list to store data
id_list = []
html = []
text_data = []
images_link = []
```

GET POST

```python
for _ in range(300): #Number of scrolling
    driver.execute_script("window.scrollBy(0, 1000);")
    # driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    print(f"Đã lướt {_+1} lần")
    time.sleep(0.1) 

# Find elements by CSS selector 
posts = driver.find_elements(By.CSS_SELECTOR, 'div.x1yztbdb.x1n2onr6.xh8yej3.x1ja2u2z')
posts_data = posts[:300]  # Limit posts

# Parsing HTML
for post in posts_data:
    post_html = post.get_attribute('outerHTML')
    soup = BeautifulSoup(post_html, 'html.parser')
    
# Extract post ID 
    post_id_tag = soup.find_all('a', href=True)
    href = post_id_tag[-1]['href']
    match = re.search(r'/posts/([^/?]+)', href)
    post_id = match.group(1)

# Extract images
    images = [img['src'] for img in soup.find_all('img', src=True) if img['src'].startswith('https://scontent')]
# Extract text
    post_data = ''
    for div in soup.find_all('div', class_="xdj266r x11i5rnm xat24cr x1mh8g0r x1vvkbs x126k92a"):
        post_data += div.text.strip() + '\n'
        

# Add to list
    id_list.append(post_id)
    html.append(post_html)
    text_data.append(post_data)
    images_link.append(images)

# Break
    if len(id_list) >= 300:
        break

driver.quit()
```

```python
df = pd.DataFrame({
    'ID': id_list,
    'HTML': html,
    'Text': text_data,
    'Image_Link': images_link
})
```

```python
df.to_csv('page_data.csv', index=False)
```

DOWNLOAD IMAGES

```python
image_folder = 'Remind'
os.mkdir(image_folder)
i = 0
# Iterate over the DataFrame rows
for index, row in df.iterrows():
    post_id = row['ID']
    image_links = row['Image_Link']  # This is a list of image URLs
    i += 1
    if image_links:  # Check if the list is not empty
        for j, img_url in enumerate(image_links):
            img_data = requests.get(img_url).content
            img_filename = os.path.join(image_folder, f'post_{i}_img_{j+1}.jpg')
            with open(img_filename, 'wb') as img_file:
                img_file.write(img_data)
            print(f"Đã tải ảnh {j+1} post {i}")
## pixel data -> wb
```
